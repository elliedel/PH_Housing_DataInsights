```python
import pandas as pd
import matplotlib.pyplot as plt

housing = pd.read_csv("datasets/processed/PH_Housing_Prices_cleaned.csv")
housing.head()
```

```python
summary_stats = housing[["floor_area_m2","price_php"]].describe()
summary_stats
```

```python
housing[["floor_area_m2","price_php"]].info()
```

```python
plt.hist(housing["price_php"]);

plt.xlabel("Price [PHP]")
plt.ylabel("Frequency")
plt.title("Distribution of Home Prices")
```

```python

plt.scatter(x=housing["floor_area_m2"], y=housing["price_php"]);

plt.xlabel("Floor Area [sq meters]")
plt.ylabel("Price [PHP]")
plt.ticklabel_format(style = 'plain')
plt.title("Price vs. Floor Area")
```

```python
housing.describe()["floor_area_m2"]
```

```python
low, high = housing["price_php"].quantile([0.1, 0.9])
mask_area = housing["price_php"].between(low,high)
housing = housing[mask_area]
```

```python
low, high = housing["floor_area_m2"].quantile([0.1, 0.9])
mask_area = housing["floor_area_m2"].between(low,high)
housing = housing[mask_area]
```

```python
housing.describe()
```

```python
# Build a scatter plot
plt.scatter(x=housing["floor_area_m2"], y=housing["price_php"]);
plt.xlabel("Floor Area [sq meters]")
plt.ylabel("Price [PHP]")
plt.ticklabel_format(style = "plain")
plt.title("Philippines: Price vs. Floor Area")
```
